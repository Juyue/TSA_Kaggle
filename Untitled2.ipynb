```python
# select fitures for data, how could you segment it into different body part?
# you could use a lot of data actually, because it is at different angles...
# then, you combine all of them, that is a great idea.
# Therefore, you first, have to train a network to do the segmentation....

# first challenge!
import read_img_tsa as tsa_rf
import os
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

# root_folder = r'E:\Juyue\Kaggle_Data\sample\sample'

root_folder = r'E:\Juyue\Kaggle_Data\sample\sample\subject_1'


def list_dir_endswith(path_name, extension_name):
    file_name_all = os.listdir(path_name)
    file_name_correct_extension = []
    for file_name in file_name_all:
        if file_name.endswith(extension_name) and not file_name.startswith('.'):
            file_name_correct_extension.append(file_name)
    
    return file_name_correct_extension

```

```python
file_extension =['.aps','.a3daps','.a3d','.ahi']

ff = 2
file_name_all = list_dir_endswith(root_folder, file_extension[ff])
data = []
for file_name in file_name_all:
    file_full_path = os.path.join(root_folder,  file_name)
    data.append(tsa_rf.read_data(file_full_path))

```

```python
# plot all first position, as well as lay them together...
%matplotlib inline

fig = plt.figure(figsize = (16,16))
for ii in range(len(file_name_all)):
    data_this = data[ii]
    ax = fig.add_subplot(4,4,ii + 1)
    im = ax.imshow(np.flipud(data_this[:,:,0].transpose()), cmap = 'viridis')
plt.show()    
```
